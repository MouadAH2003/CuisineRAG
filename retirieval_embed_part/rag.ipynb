```python
import os
import json
import docx
from tika import parser
from PyPDF2 import PdfReader
from sentence_transformers import SentenceTransformer
```

```python
def extract_text(file_path):
    # DÃ©tecter l'extension du fichier
    file_extension = os.path.splitext(file_path)[1].lower()

    #  PDF
    if file_extension == '.pdf':
        reader = PdfReader(file_path)
        text = ''
        for page in reader.pages:
            text += page.extract_text()
        return text

    #  Word
    elif file_extension == '.docx':
        doc = docx.Document(file_path)
        text = '\n'.join([para.text for para in doc.paragraphs])
        return text

    #  JSON
    elif file_extension == '.json':
        with open(file_path, 'r') as file:
            data = json.load(file)
            return json.dumps(data, indent=4)

    else:
        parsed = parser.from_file(file_path)
        return parsed['content']
```

```python
def extract_recipes(directory):
    recipes_dict = {}
    
    for filename in os.listdir(directory):
        if filename.endswith(".txt"):
            file_path = os.path.join(directory, filename)
            text = extract_text(file_path)
            recipes_dict[os.path.splitext(filename)[0]] = text
    
    return recipes_dict
```

```python
def afficher_premiers_elements(dictionnaire, nombre=5):
    for i, (cle, valeur) in enumerate(dictionnaire.items()):
        if i >= nombre:
            break
        print(f"{cle}: {valeur}...")
        print('-'*50)
```

```python
def embed_chunks(dictionnaire):
    embeddings = []
    model = SentenceTransformer('all-MiniLM-L6-v2')
    for text in dictionnaire.values():
        embedding = model.encode(text)
        embeddings.append(embedding)
    
    return embeddings
```
